# ✅ Solution: Kaiming Initialization

weight tensor 向けの **Kaiming (He) normal initialization** を実装する。

$$W \sim \mathcal{N}(0, \text{std}^2) \quad \text{where} \quad \text{std} = \sqrt{\frac{2}{\text{fan\_in}}}$$


> 💡 **どこで使う**：ReLU 系 activation を持つ層の重み初期化のデファクト。`std = sqrt(2/fan_in)` で深いネットも信号が爆発 / 消失しない。

<details>
<summary>📖 もっと詳しく</summary>

He et al. 2015。ReLU で半分の信号が消えるので fan_in/2 で補正。

対 Xavier (sigmoid/tanh 用) との違い: Xavier は `sqrt(1/fan_in)`。PyTorch `nn.Linear` のデフォルトは uniform Kaiming。Conv の場合 fan_in = `C_in × kH × kW`。

</details>

````python
import torch
import math
````

````python
# ✅ SOLUTION

def kaiming_init(weight):
    fan_in = weight.shape[1] if weight.dim() >= 2 else weight.shape[0]
    std = math.sqrt(2.0 / fan_in)
    with torch.no_grad():
        weight.normal_(0, std)
    return weight
````

````python
# Demo
import math
w = torch.empty(256, 512)
kaiming_init(w)
print(f'Mean: {w.mean():.4f} (expect ~0)')
print(f'Std:  {w.std():.4f} (expect {math.sqrt(2/512):.4f})')
````
