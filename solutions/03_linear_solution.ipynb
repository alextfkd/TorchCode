# ✅ Solution: Simple Linear Layer

全結合 linear layer を実装する: **y = xW^T + b**


> 💡 **どこで使う**：全結合層、ニューラルネットの最小単位。Transformer の Q/K/V projection も、ResNet 末尾の分類器も、結局これ。

<details>
<summary>📖 もっと詳しく</summary>

Weight shape は `(out, in)` (PyTorch 慣習)、`y = x @ W^T + b`。

Weight 初期化は Kaiming (#20) を使う。Bias は通常 0 で初期化、BatchNorm (#07) 直後なら bias は不要 (`bias=False`)。

</details>

### Signature
```python
class SimpleLinear:
    def __init__(self, in_features: int, out_features: int): ...
    def forward(self, x: torch.Tensor) -> torch.Tensor: ...
```

### Requirements
- `self.weight`: shape `(out_features, in_features)`, `randn * (1/√in_features)` で初期化
- `self.bias`: shape `(out_features,)`, zeros で初期化
- どちらも `requires_grad=True`
- `forward(x)` は `x @ W^T + b` を計算
- `torch.nn.Linear` は **使わない**

````python
import torch
import math
````

````python
# ✅ SOLUTION

class SimpleLinear:
    def __init__(self, in_features: int, out_features: int):
        self.weight = torch.randn(out_features, in_features) * (1 / math.sqrt(in_features))
        self.weight.requires_grad_(True)
        self.bias = torch.zeros(out_features, requires_grad=True)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return x @ self.weight.T + self.bias
````

````python
# Verify
layer = SimpleLinear(8, 4)
print("W shape:", layer.weight.shape)
print("b shape:", layer.bias.shape)
x = torch.randn(2, 8)
print("Output shape:", layer.forward(x).shape)
````
