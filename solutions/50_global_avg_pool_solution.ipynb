```python
# ✅ SOLUTION

def global_avg_pool(x):
    return x.mean(dim=(-2, -1))
```

```python
import torch
torch.manual_seed(0)
x = torch.randn(2, 16, 8, 8)
out = global_avg_pool(x)
print('In:', x.shape, '→ Out:', out.shape)
```
