```python
import torch
```

```python
# ✅ SOLUTION

def my_normalize(x, mean, std):
    mean = torch.as_tensor(mean, dtype=x.dtype, device=x.device)
    std = torch.as_tensor(std, dtype=x.dtype, device=x.device)
    # (C, 1, 1) に reshape — (C, H, W) と (B, C, H, W) どちらでも broadcast が効く
    shape = [-1, 1, 1]
    return (x - mean.view(shape)) / std.view(shape)
```

```python
# 🧪 Test your implementation (feel free to add more debug prints)
import torch
torch.manual_seed(0)

x = torch.rand(3, 4, 4)  # fake CIFAR-style image
mean = (0.4914, 0.4822, 0.4465)
std = (0.2470, 0.2435, 0.2616)
out = my_normalize(x, mean, std)
print('Input  shape:', x.shape, 'mean per ch:', x.mean(dim=(1,2)))
print('Output shape:', out.shape, 'mean per ch:', out.mean(dim=(1,2)))
```
