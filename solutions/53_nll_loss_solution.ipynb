```python
# ✅ SOLUTION

def my_nll_loss(log_probs, targets):
    B = log_probs.size(0)
    selected = log_probs[torch.arange(B, device=log_probs.device), targets]
    return -selected.mean()
```

```python
import torch
import torch.nn.functional as F
torch.manual_seed(0)
logits = torch.randn(4, 10)
log_probs = F.log_softmax(logits, dim=-1)
targets = torch.tensor([3, 1, 7, 0])
print('NLL: ', my_nll_loss(log_probs, targets).item())
print('Ref: ', F.nll_loss(log_probs, targets).item())
```
