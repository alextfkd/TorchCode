# ✅ Solution: Gradient Norm Clipping

**gradient norm clipping** — 学習安定化テクニック — を実装する。


> 💡 **どこで使う**：RNN / LSTM、Transformer 学習で gradient explosion を防ぐ。norm 上限を設定して超えたらスケール。

<details>
<summary>📖 もっと詳しく</summary>

`max_norm=1.0` が GPT 系の定番、ResNet 系は普通要らない。

Norm-based (向きを保ったまま縮める) が主流、value-based (各要素を clip) より安定。Loss landscape を歪めるので過度な clipping は学習を遅くする — 1.0 で適切。

</details>

### Signature
```python
def clip_grad_norm(parameters, max_norm: float) -> float:
    # 合計 norm <= max_norm になるよう gradient を in-place で clip
    # 元の (clip 前の) 合計 norm を return
```

<details>
<summary>📐 アルゴリズム手順</summary>

1. 合計 norm を計算: `sqrt(sum(p.grad.norm()^2 for p in parameters))`
2. もし合計 > max_norm: 全 grad を `max_norm / total` でスケール
3. 元の合計 norm を return

</details>

````python
import torch
````

````python
# ✅ SOLUTION

def clip_grad_norm(parameters, max_norm):
    parameters = [p for p in parameters if p.grad is not None]
    total_norm = torch.sqrt(sum(p.grad.norm() ** 2 for p in parameters))
    clip_coef = max_norm / (total_norm + 1e-6)
    if clip_coef < 1:
        for p in parameters:
            p.grad.mul_(clip_coef)
    return total_norm.item()
````

````python
# Demo
p = torch.randn(100, requires_grad=True)
(p * 10).sum().backward()
print('Before:', p.grad.norm().item())
orig = clip_grad_norm([p], max_norm=1.0)
print('After: ', p.grad.norm().item())
print('Returned:', orig)
````
