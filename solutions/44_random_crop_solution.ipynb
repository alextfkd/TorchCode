```python
import torch
import torch.nn.functional as F
```

```python
# ✅ SOLUTION

def random_crop(x, size, padding=0):
    if padding > 0:
        x = F.pad(x, [padding] * 4)
    if x.dim() == 3:
        _, H, W = x.shape
        i = torch.randint(0, H - size + 1, (1,)).item()
        j = torch.randint(0, W - size + 1, (1,)).item()
        return x[:, i:i+size, j:j+size]
    # (B, C, H, W) — per-sample offsets
    B = x.shape[0]
    H, W = x.shape[-2], x.shape[-1]
    out = torch.empty(B, x.shape[1], size, size, dtype=x.dtype, device=x.device)
    for b in range(B):
        i = torch.randint(0, H - size + 1, (1,)).item()
        j = torch.randint(0, W - size + 1, (1,)).item()
        out[b] = x[b, :, i:i+size, j:j+size]
    return out
```

```python
import torch
torch.manual_seed(0)
x = torch.rand(4, 3, 32, 32)
out = random_crop(x, size=32, padding=4)
print('In:', x.shape, '→ Out:', out.shape)
```
