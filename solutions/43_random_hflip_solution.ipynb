```python
# ✅ SOLUTION

def random_horizontal_flip(x, p=0.5):
    if x.dim() == 3:
        if torch.rand(1).item() < p:
            return x.flip(-1)
        return x
    # (B, C, H, W) — per-sample 独立な mask
    B = x.shape[0]
    mask = torch.rand(B, device=x.device) < p
    out = x.clone()
    out[mask] = x[mask].flip(-1)
    return out
```

```python
import torch
torch.manual_seed(0)
x = torch.arange(2*3*4*4).float().view(2, 3, 4, 4)
out = random_horizontal_flip(x, p=0.5)
print('Sample 0 flipped?', not torch.equal(out[0], x[0]))
print('Sample 1 flipped?', not torch.equal(out[1], x[1]))
```
