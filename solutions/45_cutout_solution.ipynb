```python
# ✅ SOLUTION

def cutout(x, size):
    out = x.clone()
    if x.dim() == 3:
        _, H, W = x.shape
        i = torch.randint(0, H - size + 1, (1,)).item()
        j = torch.randint(0, W - size + 1, (1,)).item()
        out[:, i:i+size, j:j+size] = 0
        return out
    B, _, H, W = x.shape
    for b in range(B):
        i = torch.randint(0, H - size + 1, (1,)).item()
        j = torch.randint(0, W - size + 1, (1,)).item()
        out[b, :, i:i+size, j:j+size] = 0
    return out
```

```python
import torch
torch.manual_seed(0)
x = torch.ones(1, 3, 16, 16)
out = cutout(x, size=4)
print('Zeros in output:', (out == 0).sum().item(), '(expected 3 * 4 * 4 = 48)')
```
