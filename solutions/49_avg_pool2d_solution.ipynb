```python
import torch
import torch.nn.functional as F
```

```python
# ✅ SOLUTION

def my_avg_pool2d(x, kernel_size, stride=None, padding=0):
    if stride is None:
        stride = kernel_size
    if padding > 0:
        x = F.pad(x, [padding] * 4)  # zero pad
    patches = x.unfold(2, kernel_size, stride).unfold(3, kernel_size, stride)
    return patches.mean(dim=(-1, -2))
```

```python
import torch
torch.manual_seed(0)
x = torch.arange(1*1*4*4).float().view(1, 1, 4, 4)
print('Input:\n', x.squeeze())
print('AvgPool(k=2):\n', my_avg_pool2d(x, kernel_size=2).squeeze())
```
