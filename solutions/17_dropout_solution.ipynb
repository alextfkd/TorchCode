# ✅ Solution: Dropout の実装

**Dropout** 正則化をスクラッチで実装する。


> 💡 **どこで使う**：過学習防止の古典。学習時のみ activation をランダムに 0 化、推論時は恒等。

<details>
<summary>📖 もっと詳しく</summary>

Inverted dropout (`1/(1-p)` でスケール) を採用、推論時に何もしなくて済む。

Transformer では attention dropout（attention weight に dropout）と FFN dropout を別々に設定。BatchNorm (#07) との併用は注意：BN の後に Dropout を置くと統計が壊れることがある。

</details>

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### Rules
- **training 中**: 各要素を確率 `p` で 0 にする、残りを `1/(1-p)` でスケール
- **eval 中**: input をそのまま返す (恒等)
- `nn.Dropout` や `F.dropout` は **使わない**

````python
import torch
import torch.nn as nn
````

````python
# ✅ SOLUTION

class MyDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p

    def forward(self, x):
        if not self.training or self.p == 0:
            return x
        mask = (torch.rand_like(x) > self.p).float()
        return x * mask / (1 - self.p)
````

````python
# Demo
d = MyDropout(p=0.5)
d.train()
x = torch.ones(10)
print('Train:', d(x))
d.eval()
print('Eval: ', d(x))
````
