```python
# ✅ SOLUTION

def top_k_accuracy(logits, targets, k=1):
    k = min(k, logits.size(-1))
    _, topk_indices = logits.topk(k, dim=-1)
    # topk_indices: (B, k); targets を (B, 1) に broadcast して比較
    correct = (topk_indices == targets.unsqueeze(-1)).any(dim=-1)
    return correct.float().mean().item()
```

```python
import torch
torch.manual_seed(0)
logits = torch.randn(100, 10)
targets = torch.randint(0, 10, (100,))
print(f'top-1 acc: {top_k_accuracy(logits, targets, k=1):.3f}')
print(f'top-5 acc: {top_k_accuracy(logits, targets, k=5):.3f}')
```
