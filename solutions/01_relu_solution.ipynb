# ✅ Solution: ReLU の実装

**ReLU** (Rectified Linear Unit) activation 関数をスクラッチで実装する。

$$\text{ReLU}(x) = \max(0, x)$$


> 💡 **どこで使う**：CNN / MLP の隠れ層 activation のデフォルト。負値を切るだけなので軽く、勾配消失も緩和される。

<details>
<summary>📖 もっと詳しく</summary>

AlexNet (2012) 以降 sigmoid / tanh を駆逐した定番。

弱点：`x ≤ 0` で勾配が完全に消える ("dying ReLU")。改良版に Leaky ReLU、PReLU、GELU (#19、Transformer で標準) など。

</details>

### Signature
```python
def relu(x: torch.Tensor) -> torch.Tensor:
    ...
```

### Rules
- `torch.relu`, `F.relu`, `torch.clamp` などの built-in activation は **使わない**
- autograd をサポートすること（勾配が流れる）

### Example
```
Input:  tensor([-2., -1., 0., 1., 2.])
Output: tensor([ 0.,  0., 0., 1., 2.])
```

````python
import torch
````

````python
# ✅ SOLUTION

def relu(x: torch.Tensor) -> torch.Tensor:
    return x * (x > 0).float()
````

````python
# Verify
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Output:", relu(x))
````
