# 🟢 Easy: 2D Average Pooling

**2D Average Pooling** を実装する。max pooling の smooth 版で、Global Average Pooling の構成要素でもある。

$$\text{out}[b, c, i, j] = \frac{1}{kH \cdot kW} \sum_{(p,q) \in \text{window}(i,j)} x[b, c, p, q]$$

> 💡 **どこで使う**：MaxPool の smooth 版。局所平均で down-sample、blur 効果がほしいときに。

<details>
<summary>📖 もっと詳しく</summary>

主には Global Avg Pool (#50) の形で使う、純粋な 2x2 avg pool は単体ではあまり見ない。

Padding は zero、PyTorch のデフォルト `count_include_pad=True` だと pad の 0 も平均に含まれる（端で値が小さくなる）— 通常は OK。

</details>

### MaxPool との違い
- 平均は **smooth**（どこでも微分可能）、max は non-smooth
- Padding は **zero** で行う（`-inf` ではない）、zero-pad 値は **平均に含まれる**（PyTorch default）

### Rules
- `F.avg_pool2d`, `nn.AvgPool2d`, `F.adaptive_avg_pool2d` は **使わない**
- `unfold`, `reshape`, `.mean()` は許可
- `stride` の default は `kernel_size` (non-overlapping windows)
- `padding > 0` の場合、**zero** で pad（PyTorch の `count_include_pad=True` default）
