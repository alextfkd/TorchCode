# 🟢 Easy: Per-Channel Normalize

Implement **per-channel normalization** — the very first transform in nearly every CNN training pipeline (CIFAR-10, ImageNet, ...).

$$\text{out}[c, h, w] = \frac{x[c, h, w] - \mu[c]}{\sigma[c]}$$

### Example
```
x       : (3, 32, 32) float image in [0, 1]
mean    : (0.4914, 0.4822, 0.4465)   # CIFAR-10 channel means
std     : (0.2470, 0.2435, 0.2616)   # CIFAR-10 channel stds
→ (3, 32, 32) with each channel z-scored
```

### Signature
```python
def my_normalize(x, mean, std):
    # x: (C, H, W) or (B, C, H, W)
    # mean, std: length-C tuple/list/tensor
    # Returns: same shape as x
```

### Rules
- Do **NOT** use `torchvision.transforms.Normalize` or `T.functional.normalize`
- Must support both `(C, H, W)` and `(B, C, H, W)` inputs
- Accept tuple, list, or tensor for `mean` and `std`
- Must preserve `dtype` and `device` of `x`
- Gradients must flow back to `x`
