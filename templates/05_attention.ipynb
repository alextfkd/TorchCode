# 🔴 Hard: Softmax Attention

Implement the core attention mechanism used in Transformers.

$$\text{Attention}(Q, K, V) = \text{softmax}\!\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$


> 💡 **どこで使う**：Transformer の中核演算。`softmax(QK^T/√d_k)V` — 現代モデルのほぼ全部がこの 1 式の上に乗ってる。

<details>
<summary>📖 もっと詳しく</summary>

「Attention is All You Need」(Vaswani 2017)。`√d_k` で割る正規化が無いと d が大きい時に softmax が saturate する。MultiHead (#06) や Causal Self-Attention (#09) の核として再利用される、最初に押さえる土台。

</details>
