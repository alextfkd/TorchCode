# 🟢 Easy: Weight Decay (L2 in gradient)

Implement **weight decay** — the L2 regularization term folded into the gradient before the optimizer step. This is the SGD/L2 convention; AdamW uses a different (decoupled) formulation (see problem 56).

### L2-in-gradient form (this problem)
$$g \leftarrow g + \lambda \cdot p$$

Equivalent to adding `(λ/2)·‖p‖²` to the loss and letting autograd add `λ·p` to the gradient. Doing it directly in the optimizer step skips the extra graph node.

### vs. AdamW's decoupled form
$$p \leftarrow p \cdot (1 - \text{lr} \cdot \lambda)$$
Applied to the **param** directly, not via gradient. Loshchilov & Hutter (2019) showed this works better with adaptive optimizers.

### Rules

- Skip params with `p.grad is None`
- When `weight_decay == 0`, do nothing
- Wrap in `@torch.no_grad()` — we're modifying `.grad`, not building a graph
