# 🟢 Easy: Negative Log-Likelihood Loss

**NLL Loss** を実装する。cross-entropy の後半パート。入力は **既に** `log_softmax` 済みの確率で、target クラスの log-prob を pick して負の平均を取る。

$$L = -\frac{1}{B} \sum_b \log p_b[y_b]$$

### なぜ PyTorch は log_softmax + nll_loss を分離してる
- 同じ `log_softmax` を loss と分析の両方で使い回せる
- softmax 以外から来る log-prob にも NLL を適用できる（mixture model, normalizing flow など）
- `CrossEntropyLoss(logits, targets) == nll_loss(log_softmax(logits), targets)`

```python
    # Returns: scalar loss (mean over batch)
```

### Rules
- `F.nll_loss` や `nn.NLLLoss` は **使わない**
- Input は既に `log_softmax` 済み — softmax を再計算しない
- mean reduction（scalar output）
- Advanced indexing: `log_probs[arange(B), targets]`
