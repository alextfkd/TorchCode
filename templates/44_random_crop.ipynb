# 🟢 Easy: Random Crop with Padding

**Random Crop with Zero Padding** を実装する。CIFAR-10 の定石: 4 pixel 分 zero pad してから 32×32 に random crop。

計算コストほぼゼロで学習データに平行移動 invariance を追加できる。

> 💡 **どこで使う**：CIFAR-10 学習の定石。4 pixel zero pad → 32×32 random crop で平行移動 invariance を獲得。

<details>
<summary>📖 もっと詳しく</summary>

ImageNet では `RandomResizedCrop`（scale も変える版）が標準で、より強い augmentation。

Test 時には center crop で固定。Crop と Normalize (#42) は Compose で並べて、Normalize は最後 (crop 領域の値を正規化)。

</details>

### Example
```
x: (3, 32, 32), padding=4 → (3, 40, 40) に pad → random offset で (3, 32, 32) 切り出し
```

### Rules
- `torchvision.transforms.RandomCrop` は **使わない**
- 最初に zero pad (`F.pad`) してから `size × size` で random crop
- `(C, H, W)` と `(B, C, H, W)` 両方をサポート
- batch 入力では各 sample が **独立** な offset で crop
- `size` は int 1個（正方形 output）
