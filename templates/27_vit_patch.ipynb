# 🟠 Medium: ViT Patch Embedding

Implement the **patch embedding** layer from Vision Transformer (ViT).


> 💡 **どこで使う**：Vision Transformer の入力層。画像を non-overlapping な patch に分割、各 patch を線形変換でトークン化。

<details>
<summary>📖 もっと詳しく</summary>

Dosovitskiy et al. 2020。実装は `Conv2d(in_ch, dim, kernel=patch_size, stride=patch_size)` で 1 行。`16×16 patch` が ViT-Base/Large の標準。これ + position embedding (#18) でテキストトークンと同じ形になり、以降は通常の Transformer block で処理。

</details>

### Algorithm
1. Reshape image into non-overlapping patches: `(B, C, H, W)` → `(B, N, C*P*P)`
2. Project each patch: `nn.Linear(C*P*P, embed_dim)`
3. `num_patches = (img_size // patch_size) ** 2`
