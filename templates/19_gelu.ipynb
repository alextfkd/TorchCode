# 🟢 Easy: GELU Activation

Implement the **GELU** (Gaussian Error Linear Unit) activation.

$$\text{GELU}(x) = x \cdot \Phi(x) = x \cdot 0.5 \cdot (1 + \text{erf}(x / \sqrt{2}))$$


> 💡 **どこで使う**：Transformer の活性化関数のデファクト。ReLU (#01) より滑らかで、BERT / GPT 系の標準。

<details>
<summary>📖 もっと詳しく</summary>

`x * Φ(x)` (Φ は正規分布の CDF)。`erf` 経由の正確版と tanh 近似版 (PyTorch の `approximate='tanh'`) の 2 通り。負の領域でも僅かに gradient が残るので dying ReLU 問題が緩和される。

</details>
