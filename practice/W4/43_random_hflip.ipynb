# 🟢 Easy: Random Horizontal Flip

### Example
```
x: (3, 32, 32)         → 確率 p=0.5 で flip
x: (B=8, 3, 32, 32)    → 8 sample が独立に flip
```

### Signature
```python
def random_horizontal_flip(x, p=0.5):

    # p: probability of flipping each sample
    # Returns: same shape as x
```

### Rules
- `torchvision.transforms.RandomHorizontalFlip` や `T.functional.hflip` は **使わない**
- `(C, H, W)` と `(B, C, H, W)` 両方をサポート
- batch 入力では各 sample が **独立** に flip
- `p=0` → 恒等、`p=1` → 全 sample が flip
- `dtype` と `device` を保つ
