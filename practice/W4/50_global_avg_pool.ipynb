# 🟢 Easy: Global Average Pooling

**Global Average Pooling** を実装する。feature map の空間次元を畳んで channel あたり 1 値にする。ResNet / MobileNet / EfficientNet など modern CNN classifier の標準 "head" で、大きな FC layer を置き換える。

$$\text{out}[b, c] = \frac{1}{H \cdot W} \sum_{h, w} x[b, c, h, w]$$

> 💡 **どこで使う**：現代 CNN classifier の標準 head。`(B, C, H, W) → (B, C)` で空間を畳む、FC layer の代替。

<details>
<summary>📖 もっと詳しく</summary>

ResNet, MobileNet, EfficientNet など全部これ。NIN (2013) が起源。

利点: param が大幅減 (`H×W×C` の FC weight が消える)、平行移動 invariance が built-in、入力サイズに依存しない。

</details>

### なぜ GAP
- 各 channel に global concept を持たせる
- FC head より大幅にパラメータ削減
- 平行移動 invariance が built-in


### Signature
```python
def global_avg_pool(x):
    # x: (B, C, H, W)
    # Returns: (B, C) — flattened, NOT (B, C, 1, 1)
```

### Rules
- `F.adaptive_avg_pool2d`, `nn.AdaptiveAvgPool2d`, `nn.AvgPool2d` は **使わない**
- Output shape は `(B, C)` — flatten 済みで、直接 `nn.Linear` に流せる形（`(B, C, 1, 1)` じゃない）
- 任意の `H`, `W` で動く（正方形である必要なし）
- 勾配が空間次元全体に均等に流れる
