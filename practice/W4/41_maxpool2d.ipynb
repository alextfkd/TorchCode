$$\text{out}[b, c, i, j] = \max_{(p,q) \in \text{window}(i,j)} x[b, c, p, q]$$

```python
def my_max_pool2d(x, kernel_size, stride=None, padding=0):
    # x: (B, C, H, W)
    # Returns: (B, C, H_out, W_out)
```

### Rules
- Do **NOT** use `F.max_pool2d`, `nn.MaxPool2d`, or `F.adaptive_max_pool2d`
- `unfold`, `reshape`, `torch.max`/`amax` are allowed
- `stride` defaults to `kernel_size` (non-overlapping windows)
- For `padding > 0`, pad with `float('-inf')` so negative inputs aren't dominated by the pad value
- Gradients must route only to the argmax position in each window
