> 💡 **どこで使う**：CNN の核心。kernel で画像の局所パターンを抽出する。

<details>
<summary>📖 もっと詳しく</summary>

実装は `unfold` で patch 化 → matmul（im2col アルゴリズム）。

</details>

### Rules
- `F.conv2d` や `nn.Conv2d` は **使わない**

- `F.pad` での zero-padding は許可
