# 🟢 Easy: Top-k Accuracy

Implement **top-k accuracy** — the fraction of samples where the true label is among the model's top-k predictions. Standard for ImageNet (top-1 / top-5) and most evaluation pipelines.

$$\text{acc@k} = \frac{1}{B} \sum_b \mathbb{1}[y_b \in \text{top-k}(\text{logits}_b)]$$

### Why top-k
- **top-1**: strict — what most people mean by "accuracy"
- **top-5**: more forgiving for fine-grained tasks (ImageNet has many similar dog breeds)
- top-5 is always ≥ top-1 (monotone in k)

### Rules
- Use `tensor.topk(k, dim=-1)` to get the top-k indices
- A sample is correct if `targets[b]` matches **any** of the top-k indices
- Return as a Python float (call `.item()`) — directly logable
- Clamp `k` to `num_classes` so `k > K` doesn't error (always 100% in that case)
