# 🟢 Easy: Top-k Accuracy

**Top-k accuracy** を実装する。各 sample で true label が model の top-k 予測に 含まれている割合。ImageNet (top-1 / top-5) や多くの評価 pipeline で標準。

$$\text{acc@k} = \frac{1}{B} \sum_b \mathbb{1}[y_b \in \text{top-k}(\text{logits}_b)]$$

### なぜ top-k
- **top-1**: 厳しい — "精度" と言ったら普通これ
- **top-5**: fine-grained task で許容的（ImageNet には犬種が複数あったりする）
- top-5 は常に ≥ top-1（k に対して monotone）
