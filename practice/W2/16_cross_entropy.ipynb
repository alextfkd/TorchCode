# 🟢 Easy: Cross-Entropy Loss

**Cross-entropy loss** をスクラッチで実装する。

$$\text{CE}(x, y) = -\log\frac{e^{x_y}}{\sum_j e^{x_j}}$$


> 💡 **どこで使う**：分類タスクの defacto loss。logits → softmax → 正解クラスの -log → 平均。

<details>
<summary>📖 もっと詳しく</summary>

実装は log_softmax + NLL (#53) の合成 — 数値安定のため softmax を経由せず logsumexp trick を使う。

発展: Label smoothing (#51) で過剰自信を抑える、Focal Loss でクラス不均衡対策。
