# 🟠 Medium: SGD with Momentum

Implement **SGD with Momentum** — the optimizer everyone learns first, still competitive with Adam for many vision tasks (ResNet, VGG).

### Update rule (PyTorch convention — no `(1-μ)` factor on g)
$$v_{t+1} = \mu \cdot v_t + g_t$$
$$p_{t+1} = p_t - \text{lr} \cdot v_{t+1}$$

Watch out: the classical Heavy-Ball / EMA form uses `μ*v + (1-μ)*g`. PyTorch's `torch.optim.SGD` uses **no `(1-μ)` factor**. Match PyTorch's convention so the matching test passes.


### Signature
```python
class MySGDMomentum:
    def __init__(self, params, lr, momentum=0.9): ...
    def zero_grad(self): ...
    def step(self): ...
```

### Rules
- Do **NOT** use `torch.optim.SGD`
- Implement `step()` and `zero_grad()` methods
- Velocity update: `v = momentum * v + p.grad` (PyTorch convention — no `(1-μ)` factor)
- Param update: `p -= lr * v`
- Initialize velocity buffers to zeros (one per param)
- Wrap `step()` in `@torch.no_grad()` (or manual context)
