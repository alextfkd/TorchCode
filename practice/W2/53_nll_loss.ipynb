# 🟢 Easy: Negative Log-Likelihood Loss

Implement **NLL loss** — the second half of cross-entropy. The input is **already** `log_softmax`'d probabilities; this function picks the log-prob at each target class and averages the negative.

$$L = -\frac{1}{B} \sum_b \log p_b[y_b]$$

### Why PyTorch separates `log_softmax` + `nll_loss`
- Reuse the same `log_softmax` output for both loss and analysis
- Apply NLL to log-probabilities from anywhere (mixture models, normalizing flows)
- `CrossEntropyLoss(logits, targets) == nll_loss(log_softmax(logits), targets)`

- Do **NOT** use `F.nll_loss` or `nn.NLLLoss`

- Mean reduction (scalar output)
- Use advanced indexing: `log_probs[arange(B), targets]`
